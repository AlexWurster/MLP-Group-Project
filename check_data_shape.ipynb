```python
train_data_path = '/home/co-chae/rds/hpc-work/home/camwheeler/camerons_datasets/mlp_eeg_data/train/data'

import numpy as np
import os


# Initialize an empty dictionary to store the filename: array_shape pairs
file_shapes = {}

# Loop through all files in the directory
for filename in os.listdir(train_data_path):
    # Check if the file is an .npy file
    if filename.endswith('.npy'):
        # Construct the full file path
        file_path = os.path.join(train_data_path, filename)
        # Load the numpy array
        try:
            np_array = np.load(file_path)
        except:
            print("$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$")
            print(f"filename = {filename}")
        # Get the shape of the numpy array and store it in the dictionary
        file_shapes[filename] = np_array.shape

# Print the dictionary to check
for filename, shape in file_shapes.items():
    print(f"{filename}: {shape}")
```
